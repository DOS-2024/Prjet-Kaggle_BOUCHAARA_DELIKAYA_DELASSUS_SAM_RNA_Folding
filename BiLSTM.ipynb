```python
import numpy as np
import pandas as pd

def process_dataframe_fast(df, max_len=457):
    """
    Version optimisée du traitement des séquences :
    - One-hot encoding
    - Normalisation robuste
    - Padding
    - Utilise un dictionnaire indexé plutôt que des filtres dans chaque boucle
    """
    mapping = {'A':0, 'C':1, 'G':2, 'U':3}
    factor = 1.4826
    
    # 1. Indexation des données pour accès rapide
    grouped = df.groupby(['sequence', 'experiment_type'])
    dms_dict = {}
    a3_dict = {}
    
    for (seq, exp), subdf in grouped:
        row = subdf.iloc[0]
        values = [row[f'reactivity_{i+1:04d}'] for i in range(len(seq))]
        if exp == 'DMS_MaP':
            dms_dict[seq] = np.array(values)
        elif exp == '2A3_MaP':
            a3_dict[seq] = np.array(values)
    
    unique_sequences = df['sequence'].unique()
    
    # 2. Pré-collecte des valeurs pour normalisation
    dms_values = np.concatenate([v[~np.isnan(v)] for v in dms_dict.values() if v is not None])
    a3_values  = np.concatenate([v[~np.isnan(v)] for v in a3_dict.values() if v is not None])
    
    dms_median = np.nanmedian(dms_values)
    dms_mad    = np.nanmedian(np.abs(dms_values - dms_median))
    a3_median  = np.nanmedian(a3_values)
    a3_mad     = np.nanmedian(np.abs(a3_values - a3_median))
    
    print(f"[INFO] DMS median={dms_median:.3f}, MAD={dms_mad:.3f}")
    print(f"[INFO] 2A3 median={a3_median:.3f}, MAD={a3_mad:.3f}")
    
    # 3. Création du tensor par séquence
    result_list = []
    for seq in unique_sequences:
        seq_len = len(seq)
        tensor = np.zeros((max_len, 6), dtype=np.float32)
        
        # One-hot encoding rapide
        for i, nuc in enumerate(seq):
            if i >= max_len: break
            if nuc in mapping:
                tensor[i, mapping[nuc]] = 1
        
        # Ajout réactivités normalisées
        dms_vals = dms_dict.get(seq, np.full(seq_len, np.nan))
        a3_vals  = a3_dict.get(seq, np.full(seq_len, np.nan))
        
        if len(dms_vals) > 0:
            dms_norm = np.nan_to_num((dms_vals - dms_median) / (factor * dms_mad))
            tensor[:min(seq_len, max_len), 4] = dms_norm[:max_len]
        if len(a3_vals) > 0:
            a3_norm = np.nan_to_num((a3_vals - a3_median) / (factor * a3_mad))
            tensor[:min(seq_len, max_len), 5] = a3_norm[:max_len]
        
        result_list.append(tensor)
    
    return result_list
```

```python
# 🟡 2. Transformation des trois jeux
ohe_train_data = process_dataframe_fast(df_train)
ohe_test_data  = process_dataframe_fast(df_test)
ohe_val_data   = process_dataframe_fast(df_val)

print(f"Train : {len(ohe_train_data)} séquences")
print(f"Test  : {len(ohe_test_data)} séquences")
print(f"Val   : {len(ohe_val_data)} séquences")
```

```python
import numpy as np

def split_X_Y(tensor_list):
    """
    Sépare X (features) et Y (targets) à partir d'une liste de tableaux numpy.
    Chaque tableau est de la forme (seq_len, 6) : 4 colonnes one-hot + 2 réactivités.
    
    Args:
        tensor_list (list of np.ndarray): Liste de séquences traitées.
    
    Returns:
        X: np.ndarray de shape (n_samples, seq_len, 4)
        Y: np.ndarray de shape (n_samples, seq_len, 2)
    """
    X_list = []
    Y_list = []
    
    for tensor in tensor_list:
        X_list.append(tensor[:, :4])  # 4 colonnes de one-hot
        Y_list.append(tensor[:, 4:])  # 2 colonnes de réactivités normalisées
    
    X = np.stack(X_list)
    Y = np.stack(Y_list)
    return X, Y
```

```python
X_train, Y_train = split_X_Y(ohe_train_data)
X_val, Y_val = split_X_Y(ohe_val_data)
X_test, Y_test = split_X_Y(ohe_test_data)

print("X_train:", X_train.shape)
print("Y_train:", Y_train.shape)
print("X_val:", X_val.shape)
print("Y_val:", Y_val.shape)
print("X_test:", X_test.shape)
print("Y_test:", Y_test.shape)
```

```python
import torch
import torch.nn as nn
from torch.utils.data import TensorDataset, DataLoader
import numpy as np
import matplotlib.pyplot as plt

# -------------------------------
# 1️⃣ Vérification du device
# -------------------------------
device = torch.device("mps" if torch.backends.mps.is_available() else "cpu")
print("Using device:", device)

# -------------------------------
# 2️⃣ Conversion numpy → torch
# -------------------------------
X_train_t = torch.tensor(X_train, dtype=torch.float32).to(device)
Y_train_t = torch.tensor(Y_train, dtype=torch.float32).to(device)

X_val_t = torch.tensor(X_val, dtype=torch.float32).to(device)
Y_val_t = torch.tensor(Y_val, dtype=torch.float32).to(device)

X_test_t = torch.tensor(X_test, dtype=torch.float32).to(device)
Y_test_t = torch.tensor(Y_test, dtype=torch.float32).to(device)

# -------------------------------
# 3️⃣ DataLoader
# -------------------------------
batch_size = 128
train_dataset = TensorDataset(X_train_t, Y_train_t)
train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)

val_dataset = TensorDataset(X_val_t, Y_val_t)
val_loader = DataLoader(val_dataset, batch_size=batch_size, shuffle=False)


```

```python
# -------------------------------
# 4️⃣ Masques pour ignore padding
# -------------------------------
def create_mask(y):
    # True où les valeurs sont valides
    return (y != 0).float()

# -------------------------------
# 5️⃣ Metrics masquées
# -------------------------------
def masked_mse(y_pred, y_true):
    mask = create_mask(y_true)
    mse = ((y_pred - y_true) ** 2 * mask).sum() / mask.sum()
    return mse

def masked_mae(y_pred, y_true):
    mask = create_mask(y_true)
    mae = (torch.abs(y_pred - y_true) * mask).sum() / mask.sum()
    return mae

def masked_rmse(y_pred, y_true):
    return torch.sqrt(masked_mse(y_pred, y_true))
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F

class BiLSTMOptim(nn.Module):
    def __init__(self, input_size=4, hidden1=128, hidden2=64, output_size=2, dropout=0.3):
        super(BiLSTMOptim, self).__init__()
        self.lstm1 = nn.LSTM(input_size, hidden1, batch_first=True, bidirectional=True)
        self.norm1 = nn.LayerNorm(hidden1*2)
        self.drop1 = nn.Dropout(dropout)
        
        self.lstm2 = nn.LSTM(hidden1*2, hidden2, batch_first=True, bidirectional=True)
        self.norm2 = nn.LayerNorm(hidden2*2)
        self.drop2 = nn.Dropout(dropout)
        
        # Linear pour residual
        self.residual_fc = nn.Linear(hidden1*2, hidden2*2)
        
        self.fc = nn.Linear(hidden2*2, output_size)
    
    def forward(self, x):
        out, _ = self.lstm1(x)
        out = self.norm1(out)
        out = self.drop1(out)
        
        out2, _ = self.lstm2(out)
        out2 = self.norm2(out2)
        out2 = self.drop2(out2)
        
        # Residual avec projection
        out2 += self.residual_fc(out)
        
        out = self.fc(out2)
        return out
```

```python
import torch
import torch.nn as nn

# 🔹 Paramètres du modèle
input_size = X_train.shape[2]  # nombre de features par timestep
hidden1 = 128
hidden2 = 64
output_size = Y_train.shape[2]  # nombre de targets
dropout = 0.3

# 🔹 Initialisation du modèle
device = torch.device("mps" if torch.backends.mps.is_available() else "cpu")
print(f"🔹 Entraînement sur : {device}")
model = BiLSTMOptim(input_size=input_size,
                    hidden1=hidden1,
                    hidden2=hidden2,
                    output_size=output_size,
                    dropout=dropout).to(device)

print(model)

```

```python
# -------------------------------
# 7️⃣ Optimizer & Loss
# -------------------------------
optimizer = torch.optim.Adam(model.parameters(), lr=1e-3)
criterion = nn.MSELoss(reduction='none')  # masque appliqué manuellement

# -------------------------------
# 8️⃣ Entraînement avec suivi metrics
# -------------------------------
num_epochs = 40

train_losses, val_losses = [], []
train_mses, val_mses = [], []
train_maes, val_maes = [], []
train_rmses, val_rmses = [], []

for epoch in range(num_epochs):
    model.train()
    epoch_loss = 0
    epoch_mse = 0
    epoch_mae = 0
    epoch_rmse = 0
    
    for batch_x, batch_y in train_loader:
        batch_x = batch_x.to(device)
        batch_y = batch_y.to(device)
        
        optimizer.zero_grad()
        outputs = model(batch_x)
        
        mask = create_mask(batch_y).to(device)
        loss = ((outputs - batch_y)**2 * mask).sum() / mask.sum()
        loss.backward()
        optimizer.step()
        
        # Metrics
        epoch_loss += loss.item()
        epoch_mse += masked_mse(outputs, batch_y).item()
        epoch_mae += masked_mae(outputs, batch_y).item()
        epoch_rmse += masked_rmse(outputs, batch_y).item()
    
    # Moyenne par batch
    epoch_loss /= len(train_loader)
    epoch_mse /= len(train_loader)
    epoch_mae /= len(train_loader)
    epoch_rmse /= len(train_loader)
    
    train_losses.append(epoch_loss)
    train_mses.append(epoch_mse)
    train_maes.append(epoch_mae)
    train_rmses.append(epoch_rmse)
    
    # Validation
    model.eval()
    val_loss = 0
    val_mse = 0
    val_mae = 0
    val_rmse = 0
    with torch.no_grad():
        for batch_x, batch_y in val_loader:
            batch_x = batch_x.to(device)
            batch_y = batch_y.to(device)
            
            outputs = model(batch_x)
            mask = create_mask(batch_y).to(device)
            loss = ((outputs - batch_y)**2 * mask).sum() / mask.sum()
            
            val_loss += loss.item()
            val_mse += masked_mse(outputs, batch_y).item()
            val_mae += masked_mae(outputs, batch_y).item()
            val_rmse += masked_rmse(outputs, batch_y).item()
    
    # Moyenne par batch
    val_loss /= len(val_loader)
    val_mse /= len(val_loader)
    val_mae /= len(val_loader)
    val_rmse /= len(val_loader)
    
    val_losses.append(val_loss)
    val_mses.append(val_mse)
    val_maes.append(val_mae)
    val_rmses.append(val_rmse)

    print(f"Epoch [{epoch+1}/{num_epochs}] "
          f"Train Loss: {epoch_loss:.4f}, Val Loss: {val_loss:.4f} | "
          f"Train MSE: {epoch_mse:.4f}, Val MSE: {val_mse:.4f} | "
          f"Train MAE: {epoch_mae:.4f}, Val MAE: {val_mae:.4f} | "
          f"Train RMSE: {epoch_rmse:.4f}, Val RMSE: {val_rmse:.4f}")

```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# -------------------------------
# 1️⃣ Fonction pour afficher les courbes d'entraînement
# -------------------------------
def plot_training_history_torch(train_losses, val_losses, train_maes=None, val_maes=None,
                                train_rmses=None, val_rmses=None):
    """Affiche les courbes de Loss, MAE et RMSE pendant l'entraînement."""
    n_plots = 1
    if train_maes is not None:
        n_plots += 1
    if train_rmses is not None:
        n_plots += 1

    plt.figure(figsize=(6 * n_plots, 5))

    # --- Loss ---
    plt.subplot(1, n_plots, 1)
    plt.plot(train_losses, label='Train Loss', color='blue')
    plt.plot(val_losses, label='Val Loss', color='orange')
    plt.title('Courbe de Loss (MSE)')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    plt.grid(True)

    # --- MAE ---
    if train_maes is not None:
        plt.subplot(1, n_plots, 2)
        plt.plot(train_maes, label='Train MAE', color='blue')
        plt.plot(val_maes, label='Val MAE', color='orange')
        plt.title('Courbe de MAE')
        plt.xlabel('Epoch')
        plt.ylabel('MAE')
        plt.legend()
        plt.grid(True)

    # --- RMSE ---
    if train_rmses is not None:
        plt.subplot(1, n_plots, n_plots)
        plt.plot(train_rmses, label='Train RMSE', color='blue')
        plt.plot(val_rmses, label='Val RMSE', color='orange')
        plt.title('Courbe de RMSE')
        plt.xlabel('Epoch')
        plt.ylabel('RMSE')
        plt.legend()
        plt.grid(True)

    plt.tight_layout()
    plt.show()

```

```python
# -------------------------------
# 2️⃣ Prédiction vs Réalité pour une séquence
# -------------------------------
def plot_pred_vs_true_torch(y_true, y_pred, seq_index=0, target=0):
    y_true = y_true.cpu().numpy()
    y_pred = y_pred.cpu().numpy()

    true_seq = y_true[seq_index, :, target]
    pred_seq = y_pred[seq_index, :, target]

    # masque padding
    mask = true_seq != 0
    true_seq = true_seq[mask]
    pred_seq = pred_seq[mask]

    plt.figure(figsize=(14, 5))
    plt.plot(true_seq, label='Vraie réactivité', color='green')
    plt.plot(pred_seq, label='Prédiction', color='red', alpha=0.7)
    plt.title(f"Réactivité réelle vs prédite — Séquence {seq_index} — Target {target}")
    plt.xlabel("Position")
    plt.ylabel("Réactivité")
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
# -------------------------------
# 3️⃣ Distribution globale des erreurs
# -------------------------------
def plot_error_distribution_torch(y_true, y_pred):
    y_true = y_true.cpu().numpy()
    y_pred = y_pred.cpu().numpy()

    errors = (y_pred - y_true).flatten()
    errors = errors[~np.isnan(errors)]
    errors = errors[errors != 0]

    plt.figure(figsize=(10, 5))
    plt.hist(errors, bins=100, color='purple', alpha=0.7)
    plt.title('Distribution des erreurs (y_pred - y_true)')
    plt.xlabel('Erreur')
    plt.ylabel('Fréquence')
    plt.grid(True)
    plt.show()
```

```python
# -------------------------------
# 4️⃣ Heatmap sur plusieurs séquences
# -------------------------------
def plot_heatmap_sequences_torch(y, n_sequences=30, target=0, title="Réactivité"):
    y = y.cpu().numpy()
    subset = y[:n_sequences, :, target]
    plt.figure(figsize=(15, 8))
    sns.heatmap(subset, cmap="viridis", cbar=True)
    plt.title(title)
    plt.xlabel("Position")
    plt.ylabel("Séquence")
    plt.show()
```

```python
# Courbes Loss / MAE / RMSE
plot_training_history_torch(
    train_losses, val_losses,
    train_maes, val_maes,
    train_rmses, val_rmses
)
```

```python
# 🔹 Mettre le modèle en mode évaluation
model.eval()

all_preds = []
all_trues = []

with torch.no_grad():
    for batch_x, batch_y in test_loader:
        batch_x = batch_x.to(device)
        batch_y = batch_y.to(device)
        
        outputs = model(batch_x)
        
        all_preds.append(outputs.cpu())
        all_trues.append(batch_y.cpu())

# Concaténer tous les batches
y_pred = torch.cat(all_preds, dim=0).numpy()
y_true = torch.cat(all_trues, dim=0).numpy()

# Créer le masque pour ignorer les padding
mask = create_mask(torch.tensor(y_true)).numpy()

# ✅ Corrigé : utiliser astype(bool) au lieu de .bool()
y_pred_masked = y_pred[mask.astype(bool)]
y_true_masked = y_true[mask.astype(bool)]

# 🔹 Calcul des métriques sur le jeu test
mse = np.mean((y_pred_masked - y_true_masked)**2)
mae = np.mean(np.abs(y_pred_masked - y_true_masked))
rmse = np.sqrt(mse)

print(f"Test MSE: {mse:.4f}")
print(f"Test MAE: {mae:.4f}")
print(f"Test RMSE: {rmse:.4f}")

# 🔹 Visualisation
plot_pred_vs_true_torch(torch.cat(all_trues, dim=0),
                        torch.cat(all_preds, dim=0),
                        seq_index=0, target=0)

plot_error_distribution_torch(torch.cat(all_trues, dim=0),
                              torch.cat(all_preds, dim=0))

plot_heatmap_sequences_torch(torch.cat(all_preds, dim=0),
                             n_sequences=30, target=0,
                             title="Prédictions sur Test")
```
