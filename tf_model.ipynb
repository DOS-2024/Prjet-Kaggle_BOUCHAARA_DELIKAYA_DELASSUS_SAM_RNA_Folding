```python
import numpy as np
import pandas as pd

def process_dataframe_fast(df, max_len=457):
    """
    Version optimisée du traitement des séquences :
    - One-hot encoding
    - Normalisation robuste
    - Padding
    - Utilise un dictionnaire indexé plutôt que des filtres dans chaque boucle
    """
    mapping = {'A':0, 'C':1, 'G':2, 'U':3}
    factor = 1.4826
    
    # 1. Indexation des données pour accès rapide
    grouped = df.groupby(['sequence', 'experiment_type'])
    dms_dict = {}
    a3_dict = {}
    
    for (seq, exp), subdf in grouped:
        row = subdf.iloc[0]
        values = [row[f'reactivity_{i+1:04d}'] for i in range(len(seq))]
        if exp == 'DMS_MaP':
            dms_dict[seq] = np.array(values)
        elif exp == '2A3_MaP':
            a3_dict[seq] = np.array(values)
    
    unique_sequences = df['sequence'].unique()
    
    # 2. Pré-collecte des valeurs pour normalisation
    dms_values = np.concatenate([v[~np.isnan(v)] for v in dms_dict.values() if v is not None])
    a3_values  = np.concatenate([v[~np.isnan(v)] for v in a3_dict.values() if v is not None])
    
    dms_median = np.nanmedian(dms_values)
    dms_mad    = np.nanmedian(np.abs(dms_values - dms_median))
    a3_median  = np.nanmedian(a3_values)
    a3_mad     = np.nanmedian(np.abs(a3_values - a3_median))
    
    print(f"[INFO] DMS median={dms_median:.3f}, MAD={dms_mad:.3f}")
    print(f"[INFO] 2A3 median={a3_median:.3f}, MAD={a3_mad:.3f}")
    
    # 3. Création du tensor par séquence
    result_list = []
    for seq in unique_sequences:
        seq_len = len(seq)
        tensor = np.zeros((max_len, 6), dtype=np.float32)
        
        # One-hot encoding rapide
        for i, nuc in enumerate(seq):
            if i >= max_len: break
            if nuc in mapping:
                tensor[i, mapping[nuc]] = 1
        
        # Ajout réactivités normalisées
        dms_vals = dms_dict.get(seq, np.full(seq_len, np.nan))
        a3_vals  = a3_dict.get(seq, np.full(seq_len, np.nan))
        
        if len(dms_vals) > 0:
            dms_norm = np.nan_to_num((dms_vals - dms_median) / (factor * dms_mad))
            tensor[:min(seq_len, max_len), 4] = dms_norm[:max_len]
        if len(a3_vals) > 0:
            a3_norm = np.nan_to_num((a3_vals - a3_median) / (factor * a3_mad))
            tensor[:min(seq_len, max_len), 5] = a3_norm[:max_len]
        
        result_list.append(tensor)
    
    return result_list
```

```python
import numpy as np

def split_X_Y(tensor_list):
    """
    Sépare X (features) et Y (targets) à partir d'une liste de tableaux numpy.
    Chaque tableau est de la forme (seq_len, 6) : 4 colonnes one-hot + 2 réactivités.
    
    Args:
        tensor_list (list of np.ndarray): Liste de séquences traitées.
    
    Returns:
        X: np.ndarray de shape (n_samples, seq_len, 4)
        Y: np.ndarray de shape (n_samples, seq_len, 2)
    """
    X_list = []
    Y_list = []
    
    for tensor in tensor_list:
        X_list.append(tensor[:, :4])  # 4 colonnes de one-hot
        Y_list.append(tensor[:, 4:])  # 2 colonnes de réactivités normalisées
    
    X = np.stack(X_list)
    Y = np.stack(Y_list)
    return X, Y
```

```python
# -------------------------------
# 4️⃣ Masques pour ignore padding
# -------------------------------
def create_mask(y):
    # True où les valeurs sont valides
    return (y != 0).float()

# -------------------------------
# 5️⃣ Metrics masquées
# -------------------------------
def masked_mse(y_pred, y_true):
    mask = create_mask(y_true)
    mse = ((y_pred - y_true) ** 2 * mask).sum() / mask.sum()
    return mse

def masked_mae(y_pred, y_true):
    mask = create_mask(y_true)
    mae = (torch.abs(y_pred - y_true) * mask).sum() / mask.sum()
    return mae

def masked_rmse(y_pred, y_true):
    return torch.sqrt(masked_mse(y_pred, y_true))
```

```python
import torch
import torch.nn as nn
import torch.optim as optim

# -------------------------------
# Transformer Block
# -------------------------------
class TransformerBlock(nn.Module):
    def __init__(self, embed_dim, num_heads, ff_dim, dropout=0.1):
        super().__init__()
        self.att = nn.MultiheadAttention(embed_dim, num_heads, dropout=dropout, batch_first=True)
        self.ffn = nn.Sequential(
            nn.Linear(embed_dim, ff_dim),
            nn.ReLU(),
            nn.Linear(ff_dim, embed_dim)
        )
        self.norm1 = nn.LayerNorm(embed_dim)
        self.norm2 = nn.LayerNorm(embed_dim)
        self.dropout1 = nn.Dropout(dropout)
        self.dropout2 = nn.Dropout(dropout)

    def forward(self, x, mask=None):
        attn_output, _ = self.att(x, x, x, key_padding_mask=mask)
        x = self.norm1(x + self.dropout1(attn_output))
        ffn_output = self.ffn(x)
        x = self.norm2(x + self.dropout2(ffn_output))
        return x
```

```python
# -------------------------------
# Token + Position Embedding
# -------------------------------
class TokenAndPositionEmbedding(nn.Module):
    def __init__(self, maxlen, feature_dim=4, embed_dim=64):
        super().__init__()
        self.token_proj = nn.Linear(feature_dim, embed_dim)  # projection linéaire des features
        self.pos_emb = nn.Embedding(maxlen, embed_dim)       # embedding des positions

    def forward(self, x):
        x = x.float()
        batch_size, seq_len, _ = x.size()

        token_embedding = self.token_proj(x)  # (batch, seq_len, embed_dim)
        positions = torch.arange(seq_len, device=x.device).unsqueeze(0).expand(batch_size, seq_len)
        pos_embedding = self.pos_emb(positions)  # (batch, seq_len, embed_dim)

        return token_embedding + pos_embedding
```

```python
# -------------------------------
# Transformer Model complet
# -------------------------------
class TransformerModel(nn.Module):
    def __init__(self, maxlen, embed_dim=64, num_heads=4, ff_dim=128,
                 num_blocks=2, output_dim=2, dropout=0.1):
        super().__init__()
        self.embedding = TokenAndPositionEmbedding(maxlen, feature_dim=4, embed_dim=embed_dim)
        self.blocks = nn.ModuleList([
            TransformerBlock(embed_dim, num_heads, ff_dim, dropout)
            for _ in range(num_blocks)
        ])
        self.dropout = nn.Dropout(dropout)
        self.fc = nn.Linear(embed_dim, output_dim)  # sortie par pas de temps

    def forward(self, x, mask=None):
        x = self.embedding(x)
        for block in self.blocks:
            x = block(x, mask)
        x = self.dropout(x)
        return self.fc(x)  # ✅ (batch, seq_len, output_dim)
```

```python
import torch
from tqdm import tqdm

def train_model_advanced(model, train_loader, val_loader, criterion, optimizer, device,
                         num_epochs=50, patience=5, min_delta=1e-4, scheduler_patience=3, scheduler_factor=0.5):

    scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(
        optimizer, mode='min', factor=scheduler_factor, patience=scheduler_patience
    )

    best_val_loss = float('inf')
    patience_counter = 0
    best_model_state = None

    history = {
        'train_loss': [], 'val_loss': [],
        'train_mse': [], 'val_mse': [],
        'train_mae': [], 'val_mae': [],
        'train_rmse': [], 'val_rmse': []
    }

    for epoch in range(num_epochs):
        # -------------------------------
        # 🔹 Entraînement
        # -------------------------------
        model.train()
        train_loss, train_mse, train_mae = 0.0, 0.0, 0.0

        for batch_x, batch_y in tqdm(train_loader, desc=f"Epoch {epoch+1}/{num_epochs} [Train]"):
            batch_x = batch_x.to(device).float()
            batch_y = batch_y.to(device).float()

            optimizer.zero_grad()
            outputs = model(batch_x)
            loss = criterion(outputs, batch_y)
            loss.backward()
            optimizer.step()

            # Calcul métriques
            with torch.no_grad():
                diff = outputs - batch_y
                mse = (diff ** 2).mean()
                mae = diff.abs().mean()

            train_loss += loss.item()
            train_mse += mse.item()
            train_mae += mae.item()

        n_batches = len(train_loader)
        train_loss /= n_batches
        train_mse /= n_batches
        train_mae /= n_batches
        train_rmse = train_mse ** 0.5

        # -------------------------------
        # 🔹 Validation
        # -------------------------------
        model.eval()
        val_loss, val_mse, val_mae = 0.0, 0.0, 0.0

        with torch.no_grad():
            for batch_x, batch_y in val_loader:
                batch_x = batch_x.to(device).float()
                batch_y = batch_y.to(device).float()

                outputs = model(batch_x)
                loss = criterion(outputs, batch_y)

                diff = outputs - batch_y
                mse = (diff ** 2).mean()
                mae = diff.abs().mean()

                val_loss += loss.item()
                val_mse += mse.item()
                val_mae += mae.item()

        n_val_batches = len(val_loader)
        val_loss /= n_val_batches
        val_mse /= n_val_batches
        val_mae /= n_val_batches
        val_rmse = val_mse ** 0.5

        # -------------------------------
        # 🔹 Historique et scheduler
        # -------------------------------
        history['train_loss'].append(train_loss)
        history['val_loss'].append(val_loss)
        history['train_mse'].append(train_mse)
        history['val_mse'].append(val_mse)
        history['train_mae'].append(train_mae)
        history['val_mae'].append(val_mae)
        history['train_rmse'].append(train_rmse)
        history['val_rmse'].append(val_rmse)

        scheduler.step(val_loss)

        # -------------------------------
        # 🔹 Early Stopping
        # -------------------------------
        if val_loss + min_delta < best_val_loss:
            best_val_loss = val_loss
            patience_counter = 0
            best_model_state = model.state_dict()
        else:
            patience_counter += 1
            if patience_counter >= patience:
                print(f"⏹ Early stopping triggered after {epoch+1} epochs.")
                if best_model_state is not None:
                    model.load_state_dict(best_model_state)
                break

        # -------------------------------
        # 🔹 Affichage
        # -------------------------------
        print(f"Epoch [{epoch+1}/{num_epochs}] "
              f"Train Loss: {train_loss:.4f}, Val Loss: {val_loss:.4f} | "
              f"Train MSE: {train_mse:.4f}, Val MSE: {val_mse:.4f} | "
              f"Train MAE: {train_mae:.4f}, Val MAE: {val_mae:.4f} | "
              f"Train RMSE: {train_rmse:.4f}, Val RMSE: {val_rmse:.4f}")

    print("✅ Entraînement terminé.")
    return model, history
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# -------------------------------
# Fonction pour afficher les courbes d'entraînement
# -------------------------------
def plot_training_history_torch(train_losses, val_losses, train_maes=None, val_maes=None,
                                train_rmses=None, val_rmses=None):
    # corps de la fonction inchangé

    """
    history : dictionnaire retourné par train_model_advanced
              doit contenir au moins 'train_loss' et 'val_loss'
              peut contenir 'train_mae', 'val_mae', 'train_rmse', 'val_rmse'
    """
    metrics = ['Loss', 'MAE', 'RMSE']
    fig_idx = 1
    n_metrics = 1  # au moins la Loss
    if 'train_mae' in history and 'val_mae' in history:
        n_metrics += 1
    if 'train_rmse' in history and 'val_rmse' in history:
        n_metrics += 1

    plt.figure(figsize=(6 * n_metrics, 5))
    
    # --- Loss ---
    plt.subplot(1, n_metrics, fig_idx)
    plt.plot(history['train_loss'], label='Train Loss', color='blue')
    plt.plot(history['val_loss'], label='Val Loss', color='orange')
    plt.title('Courbe de Loss (MSE)')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    plt.grid(True)
    fig_idx += 1

    # --- MAE ---
    if 'train_mae' in history and 'val_mae' in history:
        plt.subplot(1, n_metrics, fig_idx)
        plt.plot(history['train_mae'], label='Train MAE', color='blue')
        plt.plot(history['val_mae'], label='Val MAE', color='orange')
        plt.title('Courbe de MAE')
        plt.xlabel('Epoch')
        plt.ylabel('MAE')
        plt.legend()
        plt.grid(True)
        fig_idx += 1

    # --- RMSE ---
    if 'train_rmse' in history and 'val_rmse' in history:
        plt.subplot(1, n_metrics, fig_idx)
        plt.plot(history['train_rmse'], label='Train RMSE', color='blue')
        plt.plot(history['val_rmse'], label='Val RMSE', color='orange')
        plt.title('Courbe de RMSE')
        plt.xlabel('Epoch')
        plt.ylabel('RMSE')
        plt.legend()
        plt.grid(True)

    plt.tight_layout()
    plt.show()
```

```python
# -------------------------------
# 2️⃣ Prédiction vs Réalité pour une séquence
# -------------------------------
def plot_pred_vs_true_torch(y_true, y_pred, seq_index=0, target=0):
    y_true = y_true.cpu().numpy()
    y_pred = y_pred.cpu().numpy()

    true_seq = y_true[seq_index, :, target]
    pred_seq = y_pred[seq_index, :, target]

    # masque padding
    mask = true_seq != 0
    true_seq = true_seq[mask]
    pred_seq = pred_seq[mask]

    plt.figure(figsize=(14, 5))
    plt.plot(true_seq, label='Vraie réactivité', color='green')
    plt.plot(pred_seq, label='Prédiction', color='red', alpha=0.7)
    plt.title(f"Réactivité réelle vs prédite — Séquence {seq_index} — Target {target}")
    plt.xlabel("Position")
    plt.ylabel("Réactivité")
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
# -------------------------------
# 3️⃣ Distribution globale des erreurs
# -------------------------------
def plot_error_distribution_torch(y_true, y_pred, bins=100):
    """
    Affiche la distribution des erreurs (y_pred - y_true) pour PyTorch tensors.
    
    y_true : tensor (batch ou dataset)
    y_pred : tensor (batch ou dataset)
    bins   : nombre de bins pour l'histogramme
    """
    # Convertir en numpy et flatten
    y_true = y_true.detach().cpu().numpy().flatten()
    y_pred = y_pred.detach().cpu().numpy().flatten()

    # Calcul des erreurs
    errors = y_pred - y_true

    # Supprimer NaN et inf
    errors = errors[np.isfinite(errors)]

    plt.figure(figsize=(10, 5))
    sns.histplot(errors, bins=bins, kde=True, color='purple', stat='density', alpha=0.7)
    plt.title('Distribution des erreurs (y_pred - y_true)')
    plt.xlabel('Erreur')
    plt.ylabel('Densité')
    plt.grid(True)
    plt.show()
```

```python
# -------------------------------
# 4️⃣ Heatmap sur plusieurs séquences
# -------------------------------
def plot_heatmap_sequences_torch(y, n_sequences=30, target=0, title="Réactivité", annot=False, fmt=".2f"):
    """
    Affiche un heatmap des séquences pour un target donné.

    y           : tensor PyTorch [batch, seq_len, n_targets]
    n_sequences : nombre de séquences à afficher
    target      : index de la dimension target à afficher
    title       : titre du graphique
    annot       : afficher les valeurs sur le heatmap
    fmt         : format d'affichage si annot=True
    """
    # Conversion en numpy
    y = y.detach().cpu().numpy()

    # Vérification du target
    n_targets = y.shape[2]
    if target >= n_targets:
        raise ValueError(f"Target {target} est hors limite (n_targets={n_targets})")

    # Sélection des séquences
    subset = y[:n_sequences, :, target]

    plt.figure(figsize=(15, 8))
    sns.heatmap(subset, cmap="viridis", cbar=True, annot=annot, fmt=fmt)
    plt.title(title)
    plt.xlabel("Position")
    plt.ylabel("Séquence")
    plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import torch

def plot_distribution_true_vs_pred(y_true, y_pred, mask=None):
    """
    Affiche la distribution des y_true et y_pred pour chaque target (dimension de sortie).
    """
    y_true = y_true.cpu()
    y_pred = y_pred.cpu()

    if mask is None and 'create_mask' in globals():
        mask = create_mask(y_true)
    if mask is None:
        mask = torch.ones_like(y_true, dtype=torch.bool)
    else:
        mask = mask.bool()

    n_targets = y_true.shape[-1]
    plt.figure(figsize=(7 * n_targets, 5))

    for t in range(n_targets):
        # Extraction des valeurs masquées pour cette target
        y_true_t = y_true[:, :, t][mask[:, :, t]]
        y_pred_t = y_pred[:, :, t][mask[:, :, t]]

        plt.subplot(1, n_targets, t + 1)
        sns.kdeplot(y_true_t.numpy(), label='y_true', color='blue', fill=True, alpha=0.4)
        sns.kdeplot(y_pred_t.numpy(), label='y_pred', color='orange', fill=True, alpha=0.4)
        plt.title(f"Distribution Target {t}")
        plt.xlabel("Valeur")
        plt.ylabel("Densité")
        plt.legend()
        plt.grid(True, alpha=0.3)

    plt.tight_layout()
    plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import torch

def plot_scatter_true_vs_pred(y_true, y_pred, mask=None):
    """
    Scatter plot entre y_true et y_pred pour chaque target.
    """
    y_true = y_true.cpu()
    y_pred = y_pred.cpu()

    if mask is None and 'create_mask' in globals():
        mask = create_mask(y_true)
    if mask is None:
        mask = torch.ones_like(y_true, dtype=torch.bool)
    else:
        mask = mask.bool()

    n_targets = y_true.shape[-1]
    plt.figure(figsize=(7 * n_targets, 6))

    for t in range(n_targets):
        # Extraire valeurs masquées
        true_t = y_true[:, :, t][mask[:, :, t]].numpy()
        pred_t = y_pred[:, :, t][mask[:, :, t]].numpy()

        plt.subplot(1, n_targets, t + 1)
        sns.scatterplot(x=true_t, y=pred_t, s=10, alpha=0.3)
        plt.plot([true_t.min(), true_t.max()],
                 [true_t.min(), true_t.max()],
                 color='red', linestyle='--', label='y = x')
        plt.xlabel('y_true')
        plt.ylabel('y_pred')
        plt.title(f"Scatter Plot Target {t}")
        plt.legend()
        plt.grid(True, alpha=0.3)

    plt.tight_layout()
    plt.show()
```

```python
import numpy as np
from sklearn.metrics import r2_score

# y_true et y_pred sont des arrays numpy de shape (N, 457, 2)

N, L, T = y_true.shape
r2_scores = np.zeros((N, T))

for i in range(N):
    for t in range(T):
        r2_scores[i, t] = r2_score(y_true[i, :, t], y_pred[i, :, t])

# Seuil R² > 0.5
threshold = 0.5
mask_r2 = r2_scores > threshold

# Compter le nombre de séquences avec R² > 0.5
count_by_target = mask_r2.sum(axis=0)  # total pour chaque target
total_count = mask_r2.sum()            # total global

print("Nombre de séquences avec R² > 0.5 par target :", count_by_target)
print("Nombre total de séquences avec R² > 0.5 :", total_count)
```
